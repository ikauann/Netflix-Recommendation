#### Compreender qual conteúdo está disponível em diferentes países

```python
#descobrir os 5 paises com mais conteudo.
countrys = df5[df5['country'] != 0]
top_5_country = countrys['country'].value_counts().head().to_frame().reset_index()
top_5_country.rename(columns={'index': 'countrys'}, inplace=True)
ax = sns.barplot(x="countrys", y="country", data=top_5_country)
```

```python
#usa:
usa = countrys[countrys['country'] == 'United States']
usa_content = usa.type.value_counts().to_frame().reset_index()
usa_content.rename(columns={'type':'content',
                        'index':'type'}, inplace=True)

total = usa_content.sum()[1]
plt.pie(usa_content['content'], labels=usa_content['type'],
       autopct=lambda p: f'{round(p * total / 100)}');

plt.title('Usa Content');
```

```python
#india:
india_content = countrys[countrys['country'] == 'India']
india_content = india_content.type.value_counts().to_frame().reset_index()
india_content.rename(columns={'type':'content',
                        'index':'type'}, inplace=True)

total = india_content.sum()[1]
plt.pie(india_content['content'], labels=india_content['type'],
       autopct=lambda p: f'{round(p * total / 100)}');

plt.title('India Content');
```

```python
#uk:
uk_content = countrys[countrys['country'] == 'United Kingdom']
uk_content = uk_content.type.value_counts().to_frame().reset_index()
uk_content.rename(columns={'type':'content',
                        'index':'type'}, inplace=True)

total = uk_content.sum()[1]
plt.pie(uk_content['content'], labels=uk_content['type'],
       autopct=lambda p: '{:.0f}'.format(p * total / 100));

plt.title('UK Content');
```

```python
#japan:
japan_content = countrys[countrys['country'] == 'Japan']
japan_content = japan_content.type.value_counts().to_frame().reset_index()
japan_content.rename(columns={'type':'content',
                        'index':'type'}, inplace=True)

total = japan_content.sum()[1]
plt.pie(japan_content['content'], labels=japan_content['type'],
       autopct=lambda p: f'{round(p * total / 100)}');
plt.title('Japan Content');
```

```python
#korea
korea_content = countrys[countrys['country'] == 'South Korea']
korea_content = korea_content.type.value_counts().to_frame().reset_index()
korea_content.rename(columns={'type':'content',
                        'index':'type'}, inplace=True)

total = korea_content.sum()[1]

plt.pie(korea_content['content'], labels=korea_content['type'],
       autopct=lambda p: f'{round(p * total / 100)}');
plt.title('South Korea Content');
```

#### A Netflix tem mais foco em programas de TV do que filmes nos últimos anos.
**FALSE** A netflix teve foco em filmes nos ultimos 5 anos.

```python
last_5_years = df5[df5['year'] >= '2017']
last_5_years['movie_year'] = last_5_years[last_5_years['type'] == 'Movie']['type']
last_5_years['tv_show_year'] = last_5_years[last_5_years['type'] == 'TV Show']['type']
```

```python
aux = last_5_years[['type', 'year']].groupby('type').count().reset_index()
sns.barplot(x='type', y='year', hue='type', data=aux);
```

```python
tv_shows_last_5_years = last_5_years[['year', 'tv_show_year']].groupby('year').count().reset_index()
sns.barplot(x='year', y='tv_show_year', data=tv_shows_last_5_years);
plt.title('Tv Shows', fontsize=30);
```

```python
movies_last_5_years = last_5_years[['year', 'movie_year']].groupby('year').count().reset_index()
sns.barplot(x='year', y='movie_year', data=movies_last_5_years);
plt.title('Movies', fontsize=30);
```

```python
aux.plot(x='year', kind='bar');
plt.xticks(rotation=45);
```
